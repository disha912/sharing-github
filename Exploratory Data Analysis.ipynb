```python
import pandas as pd
import seaborn as sns

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/automobileEDA.csv'
df = pd.read_csv(url)

#heatmap

sns.heatmap(df.corr())
```

```python

#jointplot : Univariate/Bivariate

sns.jointplot(x= 'engine-size', y='price',data=df, kind='reg')
```

```python

#distplot: shows histogram

sns.distplot(df['price'],kde=False, bins = 5)
```

```python
#Weak Linear Relationship


sns.regplot(x= 'engine-size', y='price',data=df)
```

```python
#Categorical variables

sns.boxplot(x = 'body-style', y = 'price' , data=df )
```

```python
#Descriptive Statistical Analysis

df.describe()

#for include all

df.describe(include = ['object'])

#value_counts : this is to calculate counts of different values in a column

drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts = drive_wheels_counts.rename(columns={'drive-wheels':'drive_wheels_counts'})
drive_wheels_counts
```
